```python
import pandas as pd
```

### Retrieving data CSV files

```python
filepath = 'data/iris.csv'

# Import data
df_csv = pd.read_csv(filepath)

# print data
#print(df_csv.loc[:5])
```

```python
df_csv.drop('Id', axis=1).head()
```

```python
# Pembatas yang berbeda - file yang dipisahkan tab (.tsv)
data = pd.read_csv(filepath, sep='\t')
data.head()
```

```python
# Pembatas yang berbeda - file yang dipisahkan Spasi
data = pd.read_csv(filepath, delim_whitespace=True)
data.head()
```

```python
# Menghilangkan Nama kolom (jangan gunakan baris pertama untuk nama kolom)
data = pd.read_csv(filepath, header=None)
data.head()
```

```python
# Specify Nama kolom
data = pd.read_csv(filepath, names=['Name1', 'Name2'])
data.head()
```

```python
# Custom missing value
data = pd.read_csv(filepath, na_values=['NA', 99])
data.head()
```

```python
filejson = 'data/data.json'

# Membaca file dengan format JSON
data = pd.read_json(filejson)
data.head()
```

```python
# Membuat object 
filepath = 'data/cardio_base.xlsx'

# Membaca data dari formal Excel
data = pd.read_excel(filepath, sheet_name='Sheet1')
data.head()
```

```python
# Membuat object
filepath = 'data/cardio_base.parquet'

# Membaca data dengan format Parquet
data = pd.read_parquet(filepath)
data.head()
```
